## Do universities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university is one of the most preferred by students?

```python
#taking the data which does not contain any null values
clean_df = df[df[['Applicants total', 'Admissions total', 'Enrolled total']].notna().any(axis=1)]
clean_df
```

```python
df1 = clean_df[['Applicants total', 'Admissions total', 'Enrolled total']]
df1.head()
```

```python
df1['preferablity percent'] = df1['Enrolled total'] / df1['Admissions total']
plt.figure(figsize=(10,5))
sns.heatmap(df1.corr(),annot=True)
plt.show()
```

```python
df2 = clean_df[['Applicants total', 'Admissions total', "Admissions yield - total"]]
df2.head()
```

```python
df2['percentage of admissions'] = df2['Admissions total'] / df2['Applicants total']
plt.figure(figsize=(10,5))
sns.heatmap(df1.corr(),annot=True)
plt.show()
```

**From above we can't really conclude that universities with a high number of applications are the preferred ones by students**

##  Do students prefer universities with low tuition and fees?

```python
df4 = clean_df[['Tuition and fees, 2013-14', 'Admissions yield - total']]
df4.head()
```

```python
df4['Tuition and fees, 2013-14'].describe()
```

```python
df4['Tution Fee Slab']=pd.cut(x=df4['Tuition and fees, 2013-14'],bins=[1,10000,25000,50000],labels=['low','medium','high'])
df4
```

```python
admission=df4.groupby('Tution Fee Slab',as_index=False)['Admissions yield - total'].mean()
plt.figure(figsize=(10,8))
sns.barplot(x='Tution Fee Slab',y='Admissions yield - total',data=admission)
plt.show()
```

### from the above barplot we can see that with low tuition fee, admission rate is higher than with higher tuition fee

## Do students prefer universities from highly populated states?

```python
clean_df['enrollment_rate']=(clean_df['Enrolled total']/clean_df['Admissions total']*100).round(2)
plt.figure(figsize=(16,8))
sns.barplot(x=clean_df['State abbreviation'],y=clean_df.enrollment_rate);
plt.title('State vs Enrollment Rate')
plt.grid(axis='y')
plt.xticks(rotation=90)
plt.show()
```

**When it comes to population and enrollment rate, in general, there is not a strong tendency; however, we see that a few universities from low-population states have the highest enrollment rates. Therefore, we can not conclude that students prefer universities because they belong to crowded American states.**

## 	Do students prefer a university based on the possibility of a higher, additional academic degree in the same university?

```python
hg_degree=clean_df['Highest degree offered'].value_counts()
plt.figure(figsize=(16,8))
plt.pie(hg_degree,labels=hg_degree.index,
       autopct='%.1f%%',startangle=140)
plt.title('Highest Degree Offered')
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x='Highest degree offered',y='enrollment_rate',data=clean_df)
plt.xticks(rotation=90)
plt.show()
```

**So from the charts there alot of collage that deosn't have a higher degree and still have a high rate of admission, so we can't say that most students choose to enroll to a universties that offer a higher degree**
